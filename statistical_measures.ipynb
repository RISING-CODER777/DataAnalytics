```python
import numpy as np 
import pandas as pd 
import seaborn as sns
```

```python
data=pd.read_csv("F:\weight-height.csv")
print(data.shape)
data.head()
```

```python
#Range
max_height=np.max(data.Height)
min_height=np.min(data.Height)
print("The Largest Number is:",max_height)
print("The Smallest Number is:",min_height)
print("Range is:",(max_height-min_height))
```

```python
#variance
data_height=data.Height
data_mean=np.mean(data_height)
data_num=sum([(x-(data_mean))**2 for x in data_height])
data_var=data_num/len(data_height)
print("Variance is:",data_var)
#print(np.var(data_height))
```

```python
#standard deviation
data_std=np.sqrt(data_var)
print("Standard deviation is:",data_std)
#np.std(data_height)
```

```python
#InterQuartile Range
#data.Height.describe()
Q1=data.Height.quantile(0.25)
Q3=data.Height.quantile(0.75)
print("First Quartile is:",Q1)
print("Third Quartile is:",Q3)
```

```python
IQR=Q3-Q1
print("IQR is:",IQR)
```

```python
lower_bd=Q1-1.5*IQR
upper_bd=Q3+1.5*IQR
print("Lower Bound is:",lower_bd)
print("Upper Bound is:",upper_bd)
```

```python
#Outliers
print("The outliers are:")
data_outliers=data[(data.Height<lower_bd)| (data.Height>upper_bd) ]
print(data_outliers)
```

```python
#plot
sns.histplot(data_outliers,kde=True)
```

```python
#no outliers
data_no=data[(data.Height>lower_bd)&(data.Height<upper_bd)]
print(data_no.head())
```

```python
sns.histplot(data_no,kde=True)
```
