```python
import numpy as np
import pandas as pd
import scipy.stats as stats
```

```python
data=pd.read_csv("/content/genotype.csv")
data.head()
```

```python
#finding the length and xbar
y=data["yield"]
n=len(y)
x_sum=np.sum(y)
x_bar=x_sum/n
print(x_bar)
#x_bar=np.mean(y)
#x_bar
```

```python
mu=97.0
std_error=np.std(y)/np.sqrt(n)
t_test=abs(x_bar-mu)/std_error
print("The t_test value is:",t_test)
```

```python
tscore, pvalue = stats.ttest_1samp(y, popmean=97.0) 
print("P value is:",pvalue)                                             
```

```python
 if(pvalue<0.05):
  print("Reject Null Hypothesis!")
else:
  print("Retain Null Hypothesis!")
```
