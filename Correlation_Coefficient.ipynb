```python
import numpy as np
import pandas as pd 
from scipy.stats import rankdata
import seaborn as sns
```

```python
def person_coef(x,y):
    x_bar=np.mean(x)
    y_bar=np.mean(y)
    numer=np.sum((x-x_bar)*(y-y_bar))
    std_x=np.sqrt(np.sum((x-x_bar)**2))
    std_y=np.sqrt(np.sum((y-y_bar)**2))
    return numer/(std_x*std_y)
```

```python
def spearman_coef(x,y):
    rx,ry=rankdata(x),rankdata(y)
    return person_coef(rx,ry)
```

```python
#main
data=pd.read_csv("F:\Salary_Data.csv")
data.head()
print("Pearson Correalation coefficient is:",person_coef(data.YearsExperience,data.Salary))
print("Spearman Correalation coefficient is:",spearman_coef(data.YearsExperience,data.Salary))
```
