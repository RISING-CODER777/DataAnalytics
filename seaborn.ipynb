```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])

for col in 'xy':
    plt.hist(data[col], alpha=0.5)
```

```python
for col in 'xy':
    sns.kdeplot(data[col], shade=True)
```

```python
for col in 'xy':
    sns.distplot(data[col])
```

```python
with sns.axes_style('white'):
    sns.jointplot("x", "y", data, kind='kde')
```

```python
with sns.axes_style('white'):
    sns.jointplot("x", "y", data, kind='hex')
```

```python
sns.pairplot(data,size=2.5)
```

```python
data=sns.load_dataset('tips')
data.head()
```

```python
with sns.axes_style('ticks'):
    g=sns.factorplot("day","total_bill",data=data,kind='box')
    g.set_axis_labels("Day","Total Bill")
```

```python
with sns.axes_style('white'):
    sns.jointplot("total_bill", "tip", data=data, kind='hex')
```

```python
sns.jointplot("total_bill", "tip", data=data, kind='reg')
```

```python
with sns.axes_style('white'):
    sns.kdeplot("total_bill", "tip", data=data)
```

```python
with sns.axes_style('white'):
    sns.distplot(data["total_bill"].values)
```

```python
with sns.axes_style(style=None):
    sns.violinplot("total_bill", "tip", data=data)
```
