```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("F:\headbrain.csv")
print(data.shape)
data.head()
```

```python
X=data["Head Size(cm^3)"].values
Y=data["Brain Weight(grams)"].values
```

```python
mean_x=np.mean(X)
mean_y=np.mean(Y)
n=len(X)
```

```python
numer=0
denom=0
for i in range(n):
    numer+=(X[i]-mean_x)*(Y[i]-mean_y)
    denom+=(X[i]-mean_x)**2
m=numer/denom
c=mean_y-(m*mean_x)
print("Coefficients are",m,c)
```

```python
max_x=np.max(X)+100
min_x=np.min(X)-100
x=np.linspace(min_x,max_x,1000)
y=c+m*x
plt.plot(x,y,color='blue',label="Regression Line")
plt.scatter(X,Y,color="orange",label="Scatter Plot")
plt.legend()
plt.show()
```

```python
#RMSE
rmse=0
for i in range(n):
    y_pred=c+m*X[i]
    rmse+=(Y[i]-y_pred)**2
rmse=np.sqrt(rmse/n)
print("Root mean square error is:",rmse)
```

```python
#R2 score
ss_total=0
ss_res=0
for i in range(n):
    y_pred=c+m*X[i]
    ss_total+=(Y[i]-mean_y)**2
    ss_res+=(Y[i]-y_pred)**2
r2=1-(ss_res/ss_total)
print("R2 Score is",r2)
```
