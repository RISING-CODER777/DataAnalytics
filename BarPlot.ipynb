```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel("gender.csv.xlsx")
print(data.shape)
data.head()
```

```python
men_data=data['men'].values
women_data=data['women'].values
```

```python
plt.bar(men_data,women_data,color="blue")
plt.xlabel("Men age")
plt.ylabel("Women Age")
plt.title("Gender age")
plt.show()
```

```python
men_gdata=data[data.men>30]
women_gdata=data[data.women>30]
print(men_gdata);
print(women_gdata)
```
