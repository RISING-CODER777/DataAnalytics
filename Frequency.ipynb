```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("F:\headbrain.csv")
FreqDist=pd.DataFrame()
print(data.shape)
data.head()
```

```python

head=data["Head_Size"]
head_freq=head.value_counts(sort=False)
FreqDist["Frequency"]=head_freq
FreqDist["Head_Size"]=FreqDist.index
FreqDist=FreqDist[["Frequency","Head_Size"]]
total=head_freq.sum()
FreqDist["Relative_Frequency"]=FreqDist.Frequency/total
FreqDist.sort_index(inplace=True)
FreqDist.head()

```

```python
f,ax=plt.subplots(figsize=(10,5))
xticks=range(1,11)
plt.bar(FreqDist.Head_Size,FreqDist.Relative_Frequency)
plt.show()
```

```python
FreqDist["cumulative_Frequency"]=FreqDist.Frequency.cumsum()/total
print(FreqDist.head())
```

```python
f,ax=plt.subplots(figsize=(10,5))
xticks =range(1,11)
plt.bar(FreqDist.Head_Size,FreqDist.cumulative_Frequency)
```
