```python
import numpy as np 
import pandas as pd  
import seaborn  as sns
```

```python
data=pd.read_csv("F:\weight-height.csv")
print(data.shape)
data.head()
```

```python
data['zscore']=(data.Height-data.Height.mean())/data.Height.std()
data.head()
```

```python
sns.histplot(data.zscore,kde=True,color="b")
```

```python
data[data.zscore>3]
```

```python
data[data.zscore<-3]
```

```python
#outliers
print("The  outliers are:")
y=data[(data.zscore<-3)| (data.zscore>3)]
y
```

```python
sns.histplot(y,kde=True,color="yellow")
```

```python
#outliers
print("No outliers are:")
y=data[(data.zscore>-3)& (data.zscore<3)]
y
```

```python
sns.histplot(y,kde=True,color="violet")
```
