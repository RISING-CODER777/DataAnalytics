```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x=np.random.randn(100)
y1=5*x+9
y2=-5*x
y3=np.random.randn(100)
```

```python
plt.rcParams.update({'figure.figsize':(10,8),'figure.dpi':100})
plt.scatter(x,y1,label=f'Y1 correaltion={np.round(np.corrcoef(x,y1)[0,1],2)}')
plt.scatter(x,y2,label=f'Y2 correaltion={np.round(np.corrcoef(x,y2)[0,1],2)}')
plt.scatter(x,y3,label=f'Y3 correaltion={np.round(np.corrcoef(x,y3)[0,1],2)}')
plt.title('Correlation and Scatter Plots')
plt.legend()
plt.show()
```
