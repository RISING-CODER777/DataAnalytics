```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import matplotlib.image as img
```

```python
data=pd.read_csv("F:\cereal.csv")
print(data.shape)
data.head()
```

```python
#bar plot
x=data['mfr'].values
y=data['protein'].values
plt.bar(x,y,color='green')
plt.xlabel('mfr')
plt.ylabel('protein')
plt.title('cerels')
plt.show()
```

```python
#scatter plot
x=data['calories']
y=data['protein']
plt.scatter(x,y,color="blue")
plt.xlabel('calories')
plt.ylabel('protein')
plt.title('Scatter Plot')
plt.show()
```

```python
#line plot
plt.style.use('seaborn-whitegrid')
x=np.linspace(0,10,1000)
plt.plot(x,np.sin(x-0),color='b')
plt.plot(x,np.sin(x-1),color='#FFDD44')
plt.plot(x,np.sin(x-2),color='chartreuse')
plt.plot(x,np.sin(x-3),color='0.75')
plt.plot(x,np.sin(x-4),color='grey')
plt.plot(x,np.sin(x-5),color=(0.9,1.0,0.5,0.7))
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Line Plot')
plt.show()
```

```python
#histogram
x=data['sodium']
plt.hist(x,bins=20,alpha=0.9)
plt.xlabel('sodium')
plt.title('Histogram')
plt.show()
```
